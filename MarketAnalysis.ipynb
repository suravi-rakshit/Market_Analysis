```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_excel("SAYANI.X.xlsx")
df=pd.DataFrame(data.head())
print(df)
grouped_by=df.groupby("Itemname")["CustomerID"].sum()
print(grouped_by)
plt.bar(grouped_by.index,grouped_by.values)
plt.xlabel("Itemname",fontsize=20)
plt.ylabel("Customer",fontsize=20)
plt.show()
```
